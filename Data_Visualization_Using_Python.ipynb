## About this file

Mission ID: A unique identifier for each space mission.

Mission Name: The name assigned to the space mission.

Launch Date: The date on which the mission was launched.

Target Type: The category of the celestial body targeted by the mission.

Target Name: The specific name of the celestial body targeted by the mission.

Mission Type: The primary objective or goal of the mission.

Launch Vehicle: The type or name of the rocket or spacecraft used to launch the mission.

##PART 1
#Multiple scatter plots
This plots helps visualize the relationship between Distance from Earth (light-years) and other variables.

Mission Type vs. Distance: This plot shows the different types of missions undertaken at various distances from Earth.

Mission Duration vs. Distance: This plot explores if there's a correlation between the distance of a mission and its duration.

Mission Cost vs. Distance: This plot examines the relationship between the distance of a mission and its cost.

Scientific Yield vs. Distance: This plot investigates if missions to farther distances yield higher scientific returns.

```python
# Define the scatter plot functions
def plot_scatter(x_col, y_col, xlabel, ylabel, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(df[x_col], df[y_col])
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

# scatter plots for different varibales
plot_scatter('Distance from Earth (light-years)', 'Mission Type', 'Distance from Earth (light-years)', 'Mission Type', 'Mission Type vs. Distance from Earth')
plot_scatter('Distance from Earth (light-years)', 'Mission Duration (years)', 'Distance from Earth (light-years)', 'Mission Duration (years)', 'Mission Duration vs. Distance from Earth')
plot_scatter('Distance from Earth (light-years)', 'Mission Cost (billion USD)', 'Distance from Earth (light-years)', 'Mission Cost (billion USD)', 'Mission Cost vs. Distance from Earth')
plot_scatter('Distance from Earth (light-years)', 'Scientific Yield (points)', 'Distance from Earth (light-years)', 'Scientific Yield (points)', 'Scientific Yield vs. Distance from Earth')
```

##part 2
**Bar Chart: Number of Missions by Target**

Type This chart provides a quick overview of the most common targets for space missions.

```python
# Convert 'Target Type' to numeric representation using value_counts and then plot.
target_type_counts = df['Target Type'].value_counts()
target_type_counts.plot(kind='bar', x=target_type_counts.index, y=target_type_counts.values)
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Target Type")
plt.ylabel("Frequency")
plt.title("Number of Missions by Target Type")
plt.show()
```

##part 3
**Line Graph: Mission Success Rate Over Time**

This graph shows how the success rate of missions has changed over the years.

```python
df.plot(kind='line', x='Launch Date', y='Mission Success (%)')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Launch Date")
plt.ylabel("Mission Success (%)")
plt.title("Mission Success Rate Over Time")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

##PART 4
**Box Plot: Fuel Consumption by Launch Vehicle**

This plot compares the fuel efficiency of different launch vehicles.

###Observations from the Plot:

Starship: Appears to have the highest median fuel consumption among the four launch vehicles.

Ariane 6 and Falcon Heavy: Seem to have a similar median fuel consumption, which is lower than that of Starship.

SLS: Appears to have the lowest median fuel consumption.

IQR: Starship also seems to have the largest IQR, suggesting a wider range of fuel consumption values compared to the other launch vehicles.

```python
unique_launch_vehicles = df['Launch Vehicle'].unique()
print(unique_launch_vehicles)
```

```python
# Create the box plot
df.boxplot(column='Fuel Consumption (tons)', by='Launch Vehicle', figsize=(10, 6))
plt.xlabel("Launch Vehicle")
plt.ylabel("Fuel Consumption (tons)")
plt.title("Fuel Consumption by Launch Vehicle")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

##pat 5
Histogram: Distribution of Mission Durations

This histogram displays the range of mission durations

```python
df.plot(kind='hist', y='Mission Duration (years)', bins=20, title='Distribution of Mission Durations')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```

##part 6
**Stacked Bar Chart: Crew Size vs. Mission Type**

This chart shows the crew size distribution across different mission types.


```python

ax = sns.barplot(x='Mission Type', y='Crew Size', dodge=False, data=df)

# Set the title using plt.title()
plt.title('Crew Size vs. Mission Type')

plt.gca().spines[['top', 'right']].set_visible(False)
plt.xlabel("Mission Type")
plt.ylabel("Crew Size")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##PART 7
**Heatmap: Correlation Matrix of Numerical Variables**

This heatmap reveals correlations between variables like mission cost, duration, and fuel consumption.

```python
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
